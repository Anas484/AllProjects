```python
# Does the Season affect athletes performance
#like fall,winter,summer,spring
```

```python
df2['race_month'] = df2['Event dates'].str.split(".").str.get(1)
```

```python
df2['race_month'] = df2['race_month'].astype('int')
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x : 'Winter' if x > 11 else "Fall" if x > 8 else 'Summer' if x > 8 else 'Spring' if x > 2 else 'Winter')
```

```python
df2[df2['race_season'] == "Summer"]
```

```python
df2.groupby('race_season')['Athlete average speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```
